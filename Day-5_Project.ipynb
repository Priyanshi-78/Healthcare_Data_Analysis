```python
condition_bill = (df.groupby("Medical Condition")["Billing Amount"].mean().sort_values(ascending=False))
print(condition_bill)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x=condition_bill.index,
    y=condition_bill.values,
    palette="viridis"
)
plt.title("Average Billing Amount by Medical Condition")
plt.xlabel("Medical Condition")
plt.ylabel("Average Billing Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
insurance_bill = (df.groupby("Insurance Provider")["Billing Amount"].sum().sort_values(ascending=False))
print(insurance_bill)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(
    x=insurance_bill.index,
    y=insurance_bill.values,
    palette="rocket"
)
plt.title("Total Billing Amount by Insurance Provider")
plt.xlabel("Insurance Provider")
plt.ylabel("Total Billing Amount")
plt.xticks(rotation=90)
plt.show()
```

```python
df["Date of Admission"] = pd.to_datetime(df["Date of Admission"])
df["Discharge Date"] = pd.to_datetime(df["Discharge Date"])

df["Hospital Stay"] = (df["Discharge Date"] - df["Date of Admission"]).dt.days
avg_stay = (df.groupby("Admission Type")["Hospital Stay"].mean().sort_values(ascending=False))
print(avg_stay)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x=avg_stay.index,
    y=avg_stay.values,
    palette="Set2"
)
plt.title("Average Hospital Stay by Admission Type")
plt.xlabel("Admission Type")
plt.ylabel("Average Hospital Stay (Days)")
plt.show()
```

```python
hospital = (df["Hospital"].value_counts().head(10))
print(hospital)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x=hospital.values,
    y=hospital.index,
    palette="magma"
)
plt.title("Top 10 Hospitals by Number of Patients")
plt.xlabel("Number of Patients")
plt.ylabel("Hospital")
plt.show()
```

```python
bins = [0,20,40,60,100]
labels = ["0-20","21-40","41-60","61+"]
df["Age Group"] = pd.cut(df["Age"],bins=bins,labels=labels)

age_bill = (df.groupby("Age Group")["Billing Amount"].sum())
print(age_bill)
```

```python
plt.figure(figsize=(6,5))
sns.barplot(
    x=age_bill.index,
    y=age_bill.values,
    palette="coolwarm"
)
plt.title("Total Billing by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Billing")
plt.show()
```

```python
doctor = (df.groupby("Doctor")["Billing Amount"].sum().sort_values(ascending=False).head(10))
print(doctor)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x=doctor.values,
    y=doctor.index,
    palette="flare"
)
plt.title("Top 10 Doctors by Revenue")
plt.xlabel("Revenue")
plt.ylabel("Doctor")
plt.show()
```

```python
test = df["Test Results"].value_counts()
percentage = (test / len(df)) * 100
print(percentage)
```

```python
plt.figure(figsize=(6,6))
plt.pie(
    percentage,
    labels=percentage.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Percentage of Test Results")
plt.show()
```

```python
condition_stay = (df.groupby("Medical Condition")["Hospital Stay"].mean().sort_values(ascending=False))
print(condition_stay)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x=condition_stay.index,
    y=condition_stay.values,
    palette="Spectral"
)
plt.title("Average Hospital Stay by Medical Condition")
plt.xlabel("Medical Condition")
plt.ylabel("Average Stay")
plt.xticks(rotation=90)
plt.show()
```

```python
medication = (df.groupby(["Medical Condition","Medication"]).size().reset_index(name="Count"))
top_medication = medication.loc[medication.groupby("Medical Condition")["Count"].idxmax()]
print(top_medication)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=top_medication,
    x="Medical Condition",
    y="Count",
    palette="Set1")

plt.title("Most Prescribed Medication by Medical Condition")
plt.xticks(rotation=90)
plt.show()
```

```python
top_10=df[['Name','Billing Amount','Age','Insurance Provider']].sort_values(by='Billing Amount',ascending=False).head(10)
print(top_10)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(
    data=top_10,
    y='Billing Amount',
    x='Name',
    hue='Insurance Provider',
    palette="viridis"
)
plt.title("Top 10 Patients by Billing Amount")
plt.xlabel("Patient")
plt.ylabel("Billing Amount")
plt.xticks(rotation=90)
plt.show()
```
